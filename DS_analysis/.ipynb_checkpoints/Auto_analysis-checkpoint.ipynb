````python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
````

````python
df = pd.read_csv('corrected_dataset.csv').drop(columns='Unnamed: 0')
````

````python
df
````

##  Dataset Analysis

````python
main_df = df.drop(columns=['metro', 'vehicleConfiguration'])
main_df
````

````python
# some data graphs

plt.plot(main_df.modelDate, main_df.price)
plt.show()
````

````python
plt.bar(
    main_df.modelDate.value_counts().index, 
    main_df.modelDate.value_counts(),
    
)
plt.xticks(np.arange(main_df.modelDate.value_counts().index.min(),
                     main_df.modelDate.value_counts().index.max(),
                     10))
print("Most frequent model_year is", int(main_df.modelDate.mode()))
plt.show()
````

````python
plt.bar(
    main_df.productionDate.value_counts().index, 
    main_df.productionDate.value_counts(),
    
)
plt.xticks(np.arange(main_df.productionDate.value_counts().index.min(),
                     main_df.productionDate.value_counts().index.max(),
                     10))
print("Most frequent productionDate is", int(main_df.productionDate.mode()))
plt.show()
````

````python
main_df.query('productionDate >1995 & modelDate > 1996').describe()
````

As shown on the graphs above we can cut our dataset according to condition: 

```productionDate >1995 & modelDate > 1996```. 

So our learning data will be only for cars *younger 1996* and whose model is *younger 1995*.

````python
cutted_df = main_df.query('productionDate >1995 & modelDate > 1996')
````
