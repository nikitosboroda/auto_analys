```python
from pathlib import Path, PurePath

import pandas as pd
```

```python
path_to_datasets = Path.cwd().parents[0] / 'datasets/all_datasets'
main_dataset_folder = Path.cwd().parents[0] / 'datasets/main_dataset'

all_files = [file for file in path_to_datasets.iterdir()]
# all_files
```

```python
def file_name_parser(file_name):
    path = PurePath(file_name).name
    return path.split("_")[3]
```

```python
main_df = pd.read_csv(all_files[0]).drop(columns=['Unnamed: 0'])
main_df['city'] = file_name_parser(all_files[0])
main_df
```

```python
for file in all_files[1:]:
    new_col = file_name_parser(file)
    df = pd.read_csv(file).drop(columns=['Unnamed: 0'])
    df['city'] = new_col
    if df.empty: 
        continue
    else:
        main_df = pd.concat([main_df, df], ignore_index=True)
```

```python
main_df.to_csv('main_dataset.csv')
```
