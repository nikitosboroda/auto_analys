```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('main_dataset.csv').drop(columns='Unnamed: 0')
df
```

```python
engine_names = []
for name in df.engine_name:
    try:
        engine_names.append(name.split('\xa0')[1])
    except Exception as e:
        print(e)
        engine_names.append(name)
        continue
```

```python
# замена скрипту выше

# df["new"] = df["engine_name"].str.replace(r"^[\d\.]*\s*", "")
```

```python
len(engine_names)  # проверка на общее количество
```

```python
df.engine_name = engine_names  # замена значений
```

```python
df
```

```python
# перевод столбца engineDisplacement во float

engineDisplacements = [float(name.split()[0]) if name.split()[0]!='undefined' else 0 for name in df.engineDisplacement]
df.engineDisplacement = engineDisplacements
```

```python
enginePowers = [float(name.split()[0]) for name in df.enginePower]
df.enginePower = enginePowers
```

```python
bodyTypes = [name.split()[0] for name in df.bodyType]
df.bodyType = bodyTypes
```

```python
df.to_csv('corrected_dataset.csv')
```
